Here we drop the columns unknown names, newprice(it has around 5k missing values out of 6k),Name(it has almost 1k above columns, so if we encoding that column it will generate 1k above columns)

```python
#encoding
  #get_dummies
#get_dummies is an encoding technique in pandas
df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
df1


```

```python
#concate
 #combine df and df1
dfe=pd.concat([df,df1],axis=1)
dfe
```

```python
dfe=dfe.drop(['Unnamed: 0','Name','New_Price','Location','Fuel_Type','Transmission','Owner_Type','Fuel_Type_Electric'],axis=1)
dfe
```

```python
#string replace
dfe['Mileage']=dfe['Mileage'].str.replace('km/kg','')
dfe
```

```python
dfe['Mileage']=dfe['Mileage'].str.replace('kmpl','')
dfe['Engine']=dfe['Engine'].str.replace('CC','')
dfe['Power']=dfe['Power'].str.replace('bhp','')
dfe
```

```python
#null
dfe['Mileage']=dfe['Mileage'].str.replace('null','0')
dfe['Engine']=dfe['Engine'].str.replace('null','0')
dfe['Power']=dfe['Power'].str.replace('null','0')
dfe
```

```python
dfe.dtypes
```

```python
#converting to float
dfe['Mileage']=dfe['Mileage'].astype(float)
dfe['Engine']=dfe['Engine'].astype(float)
dfe['Power']=dfe['Power'].astype(float)
dfe
```

```python
dfe.dtypes
```

```python
dfe.isna().sum()
```

```python
#converting to float
dfe['Mileage']=dfe['Mileage'].astype('float')
dfe['Engine']=dfe['Engine'].astype('float')
dfe['Power']=dfe['Power'].astype('float')
dfe
```

```python
dfe.loc[dfe.Mileage==0,'Mileage']=np.NaN
dfe.loc[dfe.Power==0,'Power']=np.NaN
dfe
```

```python
dfe.isna().sum()
```

```python
#handling missing values
 #mileage,power,engine====>mean
 #seat======>mode
dfe['Mileage']=dfe['Mileage'].fillna(dfe['Mileage'].mean())
dfe['Engine']=dfe['Engine'].fillna(dfe['Engine'].mean())
dfe['Power']=dfe['Power'].fillna(dfe['Power'].mean())
dfe
```

```python
dfe['Seats']=dfe['Seats'].fillna(dfe['Seats'].mode()[0])
dfe
```

```python
dfe.isna().sum()
```

```python
x=dfe.drop(['Price'],axis=1)
x
```

```python
y=dfe['Price']
y
```

```python
#training ang testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#normalization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
#Model creation
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
#regerssion plot
sns.regplot(x=dfe['Year'],y=y,color='r')
```

```python
#performance evaluation
  #MAE
from sklearn.metrics import mean_absolute_error
print('Means absolute error is',mean_absolute_error(y_test,y_pred))
```

```python
 #MAPE
from sklearn.metrics import mean_absolute_percentage_error
print('Means absolute percentage error is',mean_absolute_percentage_error(y_test,y_pred))
```

```python
#mse
from sklearn.metrics import mean_squared_error
print("mean squared error is",mean_squared_error(y_test,y_pred))
```

```python
value=mean_squared_error(y_test,y_pred)
print('root mean squared error is',value**0.5)
```

```python
#r2_score
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
